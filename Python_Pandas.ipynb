**In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.**

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Rohitkumar-88/ML_Projects/main/olympics.csv'
df = pd.read_csv(url, skiprows=4)
df
```

```python
gy = df[df.NOC == 'USA']
gy = gy.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)[['Gold', 'Silver', 'Bronze']]
gy['Total'] = gy['Gold'] + gy['Bronze'] + gy['Silver']
gy.reset_index(inplace=True)
top_usa = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby(['Edition'])]
top = pd.concat(top_usa)
top

```
